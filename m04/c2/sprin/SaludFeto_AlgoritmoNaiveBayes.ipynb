## Ajuste del algoritmo Naive Bayes

### Muestreo

```python
from sklearn.model_selection import train_test_split
X = data.loc[:, data.columns != "Target"]
y = data.loc[:, data.columns == "Target"]

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.8, random_state = 0)
```

```python
print(X_train.shape, y_train.shape)
print(X_test.shape, y_test.shape)
```

### Modelización

```python
from sklearn.naive_bayes import GaussianNB
gnb = GaussianNB()

modelNB = gnb.fit(X_train, y_train.values.ravel())
y_pred_train = modelNB.predict_proba(X_train) 
y_pred_test = modelNB.predict_proba(X_test)
```

```python
print(y_pred_train.shape)
print(y_pred_test.shape)
```

### Validación

```python
from sklearn.metrics import roc_curve, auc
#
# Encapsulamos la generación de la curva ROC en una función reutilizable
#
def plot_ROC(y_train, y_test, pred_train, pred_test):
  # CURVA ROC Y AUC PARA TRAINING
  fpr_train, tpr_train, _ = roc_curve(y_train, pred_train[:, 1])
  roc_auc_train = auc(fpr_train, tpr_train)

  # CURVA ROC Y AUC PARA TEST
  fpr_test, tpr_test, _ = roc_curve(y_test, pred_test[:, 1])
  roc_auc_test = auc(fpr_test, tpr_test)

  # GRAFICA
  plt.figure()
  lw = 2

  plt.plot(fpr_train,tpr_train,color="darkgreen",lw=lw+1,label="Train ROC (AUC = %0.2f)" % roc_auc_train,)
  plt.plot(fpr_test,tpr_test,color="darkred",lw=lw+1,label="Test ROC (AUC = %0.2f)" % roc_auc_test,)

  plt.plot([0, 1], [0, 1], color="navy", lw=lw, linestyle="--")
  plt.xlim([0.0, 1.0])
  plt.ylim([0.0, 1.05])
  plt.xlabel("False Positive Rate")
  plt.ylabel("True Positive Rate")
  plt.title("ROC")
  plt.legend(loc="lower right")
  plt.show()
```

```python
plot_ROC(
    y_train=y_train,
    y_test=y_test,
    pred_train=y_pred_train,
    pred_test=y_pred_test
    )
```

### Tunning de hiperparámetros

Vamos buscar mejor combinación de *semilla* y *tamaño del conjunto de test*, que nos dé una mayor *área bajo la curva* para test.

```python
# Fuerza bruta
# Calculamos el AUC para semillas entre [0, 5000) y test_size entre [0.3, 0.9)
from sklearn.metrics import roc_auc_score
from tqdm.auto import tqdm

scores = []

for seed in tqdm(range(0, 5000)):
  for i in range(3,9):
    ts = i / 10
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = ts , random_state = seed)

    modelNB = gnb.fit(X_train, y_train.values.ravel())
    y_pred_train = modelNB.predict_proba(X_train) 
    y_pred_test = modelNB.predict_proba(X_test)

    auc_train = roc_auc_score(y_train.values.ravel(), y_pred_train[:,1])
    auc_test = roc_auc_score(y_test.values.ravel(), y_pred_test[:,1])

    scores.append((ts, seed, auc_train, auc_test))


```

```python
# Buscar los mejores hiperparámetros con estos dos objetivos en mente
# - Minimiza el sobre-ajuste
# - Maximizar el área bajo la curva con los datos de entrenamiento

df_scores = pd.DataFrame(data=scores, columns=["test_size", "seed", "auc_train", "auc_test"])

# Me invento una métrica llamada "performance"
# - performance = auc_test * (1 - abs(auc_train - auc_test))

# Esta métrica funciona del siguiente modo
# - cuanto mayor es auc_test mejor es performance
# - cuanto mayor es el sobreajuste (auc_train - auc_test) peor es performance

# Calculamos la métrica
df_scores["performance"] = df_scores.auc_test * (1 - np.abs(df_scores.auc_train - df_scores.auc_test))

# Estimamos el sobreajuste para cada test_size y semila

df_scores["overfit_est"] = np.abs(df_scores.auc_train - df_scores.auc_test)

# # Obtenemos los hiperparámetros con mejor métrica
df_scores.nlargest(5, "performance")
```

Parece que la mejor opción es separar 80% para test y 20% para entrenamiento y usar esta semilla. 

-  4445

Vamos a entrenar un modelo con uno de estos hiperparámetros

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.8 , random_state = 4445)

modelNB = gnb.fit(X_train, y_train.values.ravel())
y_pred_train = modelNB.predict_proba(X_train) 
y_pred_test = modelNB.predict_proba(X_test)

plot_ROC(y_train, y_test, y_pred_train, y_pred_test)
```

No parece haber una correlación lineal entre el *tamaño del conjunto de test* y la *semilla elegida* con las áreas bajo la curva obtenidas. 

Vamos a investigar un poco más a ver si podemos identificar algún patrón

```python
# ¿Cómo influye el tamaño de test en el rendimiento del modelo?
fig = plt.figure(figsize=(20,5))
ax1 = sns.boxplot(data=df_scores, y="auc_train", x="test_size", ax=plt.subplot(131))
ax1.set_title("Train AUC vs test_size")

ax2 = sns.boxplot(data=df_scores, y="auc_test", x="test_size", ax=plt.subplot(132))
ax2.set_title("Test AUC vs test_size")

ax3 = sns.boxplot(data=df_scores, y="performance", x="test_size", ax=plt.subplot(133))
ax3.set_title("Performance vs test_size")

plt.show()
```

Podríamos decir que, a medida que hacemos más grande el conjunto de entrenamiento:

- podemos obtener mejor AUC de entrenamiento, pero también podemos obtener valores cada vez peores, la AUC es cada vez más variable a medida que el conjunto de entrenamiento es cada vez menor.
- en cambio los valores máximos tanto para AUC de test como para "performance" (que tiene en cuenta el factor de sobre-ajuste) són ligeramente inferiores
- los valores medios en los tres casos no varian mucho, sí lo hace la dispersión

Por lo tanto la mejora parece que tiene más que ver con elegir una semilla adecuada que en cómo dividimos los conjuntos de train/test.

En estos gráficos existen *valores atípicos (outliers)* esto nos hace pensar que estos resultados van a estar muy sujetos a este dataset concreto. Hay que tener en cuenta que la variabilidad viene provocada simplemente por formas distintas de ordenar los datos.

Es decir, probablemente, el mejor modelo entrenado con estos datos, tenga un rendimiento muy distinto con otros. Puedce que sea una buena idea descartar estas *áreas atípcas* para elegir unos buenos hiperparámetros que fucionen bien con nuevos datos.

```python
# ¿Cómo influye la semilla  en el rendimiento del modelo?

# Buscamos a ver si hay algún patrón que nos permita elegir una semilla
fig = plt.figure(figsize=(20,5))
ax1 = sns.scatterplot(data=df_scores, y="auc_train", x="seed", ax=plt.subplot(131))
ax1.set_title("Train AUC vs seed")

ax2 = sns.scatterplot(data=df_scores, y="auc_test", x="seed", ax=plt.subplot(132))
ax2.set_title("Test AUC vs seed")

ax3 = sns.scatterplot(data=df_scores, y="performance", x="seed", ax=plt.subplot(133))
ax3.set_title("Performance vs seed")

plt.show()
```
