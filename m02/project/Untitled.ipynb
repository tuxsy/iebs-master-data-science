```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
housedata = pd.read_csv("housedata.csv")
mean_price_by_city = housedata.replace(0, np.nan).groupby("city")["price"].mean().to_frame()
housedata.price = housedata.apply(
    lambda x: mean_price_by_city.loc[x.city].price if x.price == 0.0 else x.price,
    axis=1
)
```
